```python
df1 = pd.read_csv('data.csv')
df1
```

```python
df1['corrected_class'] = df1['class'].apply(lambda x : x.split("'")[1])
df1
```

```python
df2  =df1.drop('class', axis = 'columns')
df2
```

```python
input_set = df2.drop('corrected_class', axis = 'columns')
input_set
```

```python
target_set = df2.corrected_class
target_set
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,x,Y,y = train_test_split(input_set, target_set, train_size=0.7)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X,Y)
```

```python
model.score(x,y)
```
